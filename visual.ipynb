```python
with open("out.txt", "r") as file:
    data = file.read()
    data = data.replace("[", "").replace("]", "").split("\n")
    data.pop()
data = [list(map(int, line.strip("").split(", "))) for line in data]
```

```python
import matplotlib.pyplot as plt

merge_sort = [item[1] for item in data]
quick_sort = [item[2] for item in data]
heap_sort = [item[3] for item in data]
time_values = [item[0] for item in data]

plt.figure(figsize=(10, 6))
plt.plot(time_values, merge_sort, label="Merge Sort", marker="o")
plt.plot(time_values, quick_sort, label="Quick Sort", marker="o")
plt.plot(time_values, heap_sort, label="Heap Sort", marker="o")


plt.xlabel("Dataset size")
plt.ylabel("Time (s)")
plt.title("Sorting comparison")
plt.legend()
plt.grid(True)
plt.show()
```
